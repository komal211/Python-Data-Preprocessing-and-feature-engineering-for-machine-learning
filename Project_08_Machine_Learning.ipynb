## Data Cleaning and Creating additional Features

```python
df.info()
```

```python
#remove all rows or districts with at least one missing value
#there are around 100 districts with missing values
df.dropna(inplace = True)
```

```python
#house size
#larger the size, higher the value
#creating new column rooms_per_household
#dividing total rooms by number of households
df["rooms_per_household"] = df.total_rooms.div(df.households)
```

```python
#get 10 largest values
df.rooms_per_household.nlargest(10)
#1914 households whose rooms are 141
```

```python
df.rooms_per_household.nsmallest(10)
#5916 household has 0.8 room
#it does not make sense since it is not possible to have less than one room per household
```

```python
#deciding not to remove these extreme or suspicious values
df.loc[[1979, 5916, 8219]]
```

```python
#population divided by number of households
df["pop_per_household"] = df.population.div(df.households)
```

```python
#total bedrooms divided by total rooms
df["bedrooms_per_room"] = df.total_bedrooms.div(df.total_rooms)
```

## Feature Engineering - Part 1

```python
#before creating machine learning model, some features need to be transformed
#seperate features and labels
```

```python
label = df.median_house_value.copy()
label
```

```python
features = df.drop(columns = ["median_house_value"])
features
#features does not contain column median_house_value
```

```python
features.info()
```

```python
#many machine learning models do not work well with different scales
#standarize numerical features by calculating set values
features.select_dtypes("float")
```

```python
import scipy.stats as stats
```

```python
#calculate set scores
feat1 = features.select_dtypes("float").apply(lambda x: stats.zscore(x))
feat1
```

```python
pd.options.display.float_format = '{:.2f}'.format
#2 decimal point
```

```python
feat1
```

```python
feat1.agg(["mean", "std"])
#mean value in each and every column is zero 
#standard deviation in each and every column is 1
#this is the definiton of standardizing features
#all features now have same scale
```

## Feature Engineering - Part 2

```python
#machine learning algorithms cannot handle text data and transform text like categorical data into numbers
features.ocean_proximity
```

```python
features.ocean_proximity.value_counts()
#five different catergories
#most frequent - <1H OCEAN
```

```python
#create new feature column with binary values
dummies = pd.get_dummies(features.ocean_proximity)
dummies
```

```python
#for each and every district and for each and every row, exactly one time of value 1 and 4 times value 0
#if all five features are kept, one is actually redundant
#once value in four columns is known, fifth column value can be known automatically
#if working with all five dummy variables
#information twice in data and it can cost multicolinearity problem
#some algorithms aare sensitive and can be harmed by multicolinearity like linear regression
#if use linear regression, need to delete one feature and only use K-1
#random forest regression, keep all five or all key features in dataset
#merge features into one feature dataframes
#horizontally concatenation
features = pd.concat([feat1, dummies, df.income_cat], axis = 1)
features
```

## Splitting the Data into Train and Test Set

```python
#80% of data - train set
#20% of data - test set
#this division of data depends on the size of dataset
#millions of data points, 1% or 2% test set is enough
features
```

```python
#declare test set size 20% is reasonable since there are 20000 observations
test_size = 0.2
```

```python
#to split dataset by drawing a random sample using pandas sample method
#random sampling ensures reproducibility
X_test = features.sample(frac = test_size, random_state= 123)
```

```python
X_test
#random sampling drawback - rick of introducing sampling bias
#sample is not representative of the whole dataset
#median_income - most important feature
#test set contains all high income districts
#training set contains all low income districts
#could lead to poor model
#five income categories in training and test data set have same ratios as in whole dataset
```

```python
#check whether close to perfectly stratified to samples or not
#compare the relative frequencies of income categories in test set
X_test.income_cat.value_counts(normalize = True)
#actually equal or at least very close to ratios in whole dataset
```

```python
features.income_cat.value_counts(normalize = True) #whole dataset
```

```python
#it can be concluded that simply random sampling with pandas can be used
X_test.index#getting row labels
```

```python
#create training set
#check for each and every row laebl and features, whether row labels are already in test set
#if row label is not in test set, put into new dataframe X_train
X_train = features.loc[~features.index.isin(X_test.index)].copy()
```

```python
X_train
```

```python
X_train.income_cat.value_counts(normalize = True)
```

```python
#shuffle training set by using sample method from pandas
X_train = X_train.sample(frac = 1, random_state= 123)
X_train
```

```python
X_train.drop(columns = ["income_cat"], inplace = True)
```

```python
X_test.drop(columns = ["income_cat"], inplace = True)
```

```python
y_train = label.loc[X_train.index]
y_test = label.loc[X_test.index]
```

```python
y_train
```

## Training the ML Model (Random Forest Regressor)

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest_reg = RandomForestRegressor(random_state = 42, n_estimators = 500,
                                   max_features= "sqrt", max_depth= 75, min_samples_split=2)
```

```python
forest_reg.fit(X_train, y_train)
```

```python
#two metrics to measure the fit
#coefficient of determination
#mean-squared-error
forest_reg.score(X_train, y_train)
#might indiccate regressor is overfitting
```

```python
from sklearn.metrics import mean_squared_error
#lower the mean_squared error, better the fit
```

```python
#making predictions on training set
pred = forest_reg.predict(X_train)
pred
```

```python
#compared actual labels with the predicted labels
#taking square root of mean squared error
forest_mse = mean_squared_error(y_train, pred)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
#model fits training data well but does not indicate whether model generalize to new data and can predit house prices for new data
```

## Evaluating the Model on the Test Set

```python
forest_reg
```

```python
forest_reg.score(X_test, y_test) #coefficient of determination
```

```python
pred = forest_reg.predict(X_test) #X-test is the features
pred
```

```python
forest_mse = mean_squared_error(y_test, pred)#measure mean squared error based on actual values and predicted values
forest_rmse = np.sqrt(forest_mse) #root mse based on mse calculated before
forest_rmse
```

```python
#compare predicted values with actual values
comp = pd.DataFrame(data = {"True_V": y_test, "Pred":pred})
comp
```

```python
#for some observations for some district, difference is pretty high
#calculate difference between predicted and actual values using numberical absolute reference
ae = comp.True_V.sub(comp.Pred).abs()
ae
```

```python
mae = ae.mean()#calculate mean absolute difference

mae
```

```python
#on average, there is difference of over 30000 between predicted and actual value
#conclusion of explanatory data analysis - income might be the most important factor followed by the location
#random forest regressor give us the relative importance of features for the model
```
